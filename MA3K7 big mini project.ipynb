```python
import random
```

```python
def attempt(a,b,c,n):
    if n==0:
        print(a,b,c)
    elif (a==0 and b==0) or (b==0 and c==0) or (a==0 and c==0):
        attempt(a,b,c,0)
    elif a==0:
        attempt(a+2,b-1,c-1,n-1)
    elif b==0:
        attempt(a-1,b+2,c-1,n-1)
    elif c==0:
        attempt(a-1,b-1,c+2,n-1)
    else:
        d=random.randint(1,3)
        if d==1:
            attempt(a+2,b-1,c-1,n-1)
        elif d==2:
            attempt(a-1,b+2,c-1,n-1)
        elif d==3:
            attempt(a-1,b-1,c+2,n-1)
```

```python
def attempt1(a,b,c):
    while not ((a==0 and b==0) or (b==0 and c==0) or (a==0 and c==0)):
        if a==0:
            a+=2
            b-=1
            c-=1
        elif b==0:
            a-=1
            b+=2
            c-=1
        elif c==0:
            a-=1
            b-=1
            c+=2
        else:
            d=random.randint(1,3)
            if d==1:
                a+=2
                b-=1
                c-=1
            elif d==2:
                a-=1
                b+=2
                c-=1
            elif d==3:
                a-=1
                b-=1
                c+=2
    print(a,b,c)
```
