```python
def check30(a):
    b=0
    c=[]
    for i in range(3):
        for j in range(2):
            if i!=0 and (i+j+1)%3!=0:
                k=0
            if i!=1 and (i+j+1)%3!=1:
                k=1
            if i!=2 and (i+j+1)%3!=2:
                k=2
            if a[0,i]==1 and a[1,(i+j+1)%3]==1 and a[2,k]==1:
                return 0
            if a[0,i]==1 and a[1,(i+j+1)%3]==1 and a[2,k]==None:
                c.append((2,k))
            if a[0,i]==1 and a[1,(i+j+1)%3]==None and a[2,k]==1:
                c.append((1,(i+j+1)%3))
            if a[0,i]==None and a[1,(i+j+1)%3]==1 and a[2,k]==1:
                c.append((0,i))
            if a[0,i]!=None and a[1,(i+j+1)%3]!=None and a[2,k]!=None:
                b+=0
    if b==6:
        return 0
    if c==[]:
        for i in range(3):
            for j in range(3):
                if a[i,j]==None:
                    c.append((i,j))
    return c
```

```python
def check31(a):
    b=0
    c=[]
    for i in range(6):
        if i<3:
            if a[i,0]==0 and a[i,1]==0 and a[i,2]==0:
                return 0
            if a[i,0]==0 and a[i,1]==0 and a[i,2]==None:
                c.append((i,2))
            if a[i,0]==0 and a[i,1]==None and a[i,2]==0:
                c.append((i,1))
            if a[i,0]==None and a[i,1]==0 and a[i,2]==0:
                c.append((i,0))
            if a[i,0]!=None and a[i,1]!=None and a[i,2]!=None:
                b+=1
        else:
            if a[0,i-3]==0 and a[1,i-3]==0 and a[2,i-3]==0:
                return 0
            if a[0,i-3]==0 and a[1,i-3]==0 and a[2,i-3]==None:
                c.append((2,i-3))
            if a[0,i-3]==0 and a[1,i-3]==None and a[2,i-3]==0:
                c.append((1,i-3))
            if a[0,i-3]==None and a[1,i-3]==0 and a[2,i-3]==0:
                c.append((0,i-3))
            if a[0,i-3]!=None and a[1,i-3]!=None and a[2,i-3]!=None:
                b+=1
    if b==6:
        return 0
    if c==[]:
        for i in range(3):
            for j in range(3):
                if a[i,j]==None:
                    c.append((i,j))
    return c
```

```python
d=numpy.matrix([[None,None,None],[None,None,None],[None,None,None]])
```

```python
e=[]
```

```python
def cons(a,b):
    if b==0:
        f=check30(a)
        if f==0:
            e.append(a)
        else:
            for i in f:
                c=a.copy()
                c[i]=0
                cons(c,(b+1)%2)
    if b==1:
        f=check31(a)
        if f==0:
            e.append(a)
        else:
            for i in f:
                c=a.copy()
                c[i]=1
                cons(c,(b+1)%2)
```

```python
g=[d]
for i in numpy.array(e):
    for j in g:
        if (i==j).all():
            break
    else:
        g.append(i)
```

```python
h=[]
for i in range(len(g)):
    for j in range(i+1,len(g)):
        if (g[i].T==g[j]).all():
            h.append(i)
```
